```python
from keras.applications.vgg16 import VGG16
model = VGG16()
model.summary()
```

```python
import requests
import cv2
import numpy
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd
```

```python
img = []

img.append("gambar/pohon.jpg")
img.append("gambar/basketball.jpg")
img.append("gambar/golf ball.jpg")
img.append("gambar/croquet ball.jpg")
img.append("gambar/rugby ball.jpg")
img.append("gambar/cicak.jpg")
img.append("gambar/motor.jpg")
img.append("gambar/tennis ball.jpg")
img.append("gambar/soccer ball.jpg")
img.append("gambar/baseball.jpg")
img.append("gambar/kelereng.jpg")
img.append("gambar/ping ping ball.jpeg")
```

```python
imgs = []
plt.figure(figsize=(15,4))
for u in range(0,len(img)): #dari 0 sampai n
    gmbr = numpy.array(Image.open(img[u])) #konversi ke np array
    plt.subplot(1,len(img),u+1) 
    plt.imshow(gmbr) #menampilkan ke matplotlib
    gmbr = cv2.resize(gmbr, (224,224)) #resize sesuai input vgg16 yaitu 224,224
    imgs.append(gmbr) #kumpulkan semua image yang telah diproses ke imgs
plt.show() #manmpilkan ke matplotlib
imgs = numpy.array(imgs) #konersi ke np array
print(imgs.shape)
```

```python
yh = model.predict(imgs)

for i in range(len(img)):
    if (numpy.argmax(yh[i]) == 429):
        print("Ini adalah baseball")
    elif (numpy.argmax(yh[i]) == 430):
        print("Ini adalah basketball")
    elif (numpy.argmax(yh[i]) == 522):
        print("Ini adalah croquetball")
    elif (numpy.argmax(yh[i]) == 722):
        print("Ini adalah ping pong ball")
    elif (numpy.argmax(yh[i]) == 768):
        print("Ini adalah rugby ball")
    elif (numpy.argmax(yh[i]) == 805):
        print("Ini adalah soccer ball")
    elif (numpy.argmax(yh[i]) == 852):
        print("Ini adalah tennis ball")
    else:
        print("Tidak ada kelas yang sesuai")
```
